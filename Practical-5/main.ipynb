```python
import matplotlib.pyplot as plt
import csv

# Lists to store data
fib_numbers = []
iter_steps = []
rec_steps = []

# Read data from CSV file
with open('fibonacci_steps_comparison.txt', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        fib_numbers.append(int(row['Fibonacci Number']))
        iter_steps.append(int(row['Iterative Steps']))
        rec_steps.append(int(row['Recursive Steps']))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(fib_numbers, iter_steps, marker='o', linestyle='-', color='blue', label='Iterative Steps')
plt.plot(fib_numbers, rec_steps, marker='s', linestyle='--', color='red', label='Recursive Steps')

# Labels and title
plt.xlabel('Fibonacci Number (n)')
plt.ylabel('Steps Count')
plt.title('Performance Comparison: Iterative vs Recursive Fibonacci')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("fibonacci_comparison.png", dpi = 300, bbox_inches = "tight")
plt.show()
```
