```python
import matplotlib.pyplot as plt
import pandas as pd

# Function to plot chart from a given file
def plot_case(filename, case_title):
    # Read the text file as a CSV
    df = pd.read_csv(filename)

    # Plot
    plt.plot(df["Input Size"], df["Linear Steps"], marker='o', label="Linear Search", color='red')
    plt.plot(df["Input Size"], df["Binary Steps"], marker='s', label="Binary Search", color='blue')

    plt.title(f"{case_title} - Linear vs Binary Search")
    plt.xlabel("Input Size")
    plt.ylabel("Steps Executed")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{case_title}.png", dpi=300, bbox_inches="tight")
    plt.show()

# Plot for Best Case
plot_case("best_case.txt", "Best Case")

# Plot for Average Case
plot_case("average_case.txt", "Average Case")

# Plot for Worst Case
plot_case("worst_case.txt", "Worst Case")
```
