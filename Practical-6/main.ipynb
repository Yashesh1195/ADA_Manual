```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionary of data types and corresponding file names
cases = {
    "Random Data": "random.txt",
    "Ascending Data": "ascending.txt",
    "Descending Data": "descending.txt"
}

# Plotting each case in a separate figure
for case_name, filename in cases.items():
    # Read the data from file
    df = pd.read_csv(filename)
    
    # Create a new figure for each case
    plt.figure(figsize=(7, 5))
    plt.plot(df['Input'], df['BasicQuickSteps'], marker='o', linestyle='-', color='blue', label='Basic Quick Sort')
    plt.plot(df['Input'], df['RandomQuickSteps'], marker='s', linestyle='--', color='green', label='Randomized Quick Sort')

    plt.title(f'Quick Sort Comparison - {case_name}')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Steps')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"{case_name}.png", dpi=300, bbox_inches="tight")
    plt.show()
```
