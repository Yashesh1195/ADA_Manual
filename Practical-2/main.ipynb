```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# File names and titles
data_files = {
    'random.txt': 'Random Data',
    'ascending.txt': 'Ascending Data',
    'descending.txt': 'Descending Data'
}

# Loop through each data file
for file_name, title in data_files.items():
    sizes = []
    selection, bubble, insertion, merge, quick = [], [], [], [], []

    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # Skip header
        for row in reader:
            sizes.append(int(row[0]))
            selection.append(int(row[1]))
            bubble.append(int(row[2]))
            insertion.append(int(row[3]))
            merge.append(int(row[4]))
            quick.append(int(row[5]))

    # Create a numpy array for the x-axis positions
    x = np.arange(len(sizes))
    bar_width = 0.15

    # Create figure and axes
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot for each dataset
    bars1 = ax.bar(x - 2*bar_width, selection, width=bar_width, label='Selection Sort')
    bars2 = ax.bar(x - bar_width, bubble, width=bar_width, label='Bubble Sort')
    bars3 = ax.bar(x, insertion, width=bar_width, label='Insertion Sort')
    bars4 = ax.bar(x + bar_width, merge, width=bar_width, label='Merge Sort')
    bars5 = ax.bar(x + 2*bar_width, quick, width=bar_width, label='Quick Sort')

    # Add algorithm names above bars with more spacing
    for bars, name in zip([bars1, bars2, bars3, bars4, bars5],
                          ["Selection", "Bubble", "Insertion", "Merge", "Quick"]):
        for rect in bars:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2,   # center of bar
                    height + (0.02 * max(selection + bubble + insertion + merge + quick)),  # extra space
                    name, ha='center', va='bottom', fontsize=8, rotation=90)

    # # Add algorithm names above bars with more spacing (HORIZONTAL text)
    # for bars, name in zip([bars1, bars2, bars3, bars4, bars5],
    #                       ["Selection", "Bubble", "Insertion", "Merge", "Quick"]):
    #     for rect in bars:
    #         height = rect.get_height()
    #         ax.text(rect.get_x() + rect.get_width()/2,   # center of bar
    #                 height + (0.02 * max(selection + bubble + insertion + merge + quick)),  # extra space
    #                 name, ha='center', va='bottom', fontsize=8)  # <-- removed rotation=90


    # Title and labels
    ax.set_title(f'Input Size vs Steps Count ({title})')
    ax.set_xlabel('Input Size (n)')
    ax.set_ylabel('Steps Count')
    ax.set_xticks(x)
    ax.set_xticklabels(sizes)

    ax.legend()
    ax.grid(True, axis='y', linestyle="--", alpha=0.7)

    plt.tight_layout()
    plt.savefig(f"graph_{title}.png", dpi=300, bbox_inches="tight")
    plt.show()
```
