```python
import matplotlib.pyplot as plt
import csv

# File names and titles
data_files = {
    'random.txt': 'Random Data',
    'ascending.txt': 'Ascending Data',
    'descending.txt': 'Descending Data'
}

# Loop through each data file
for file_name, title in data_files.items():
    sizes = []
    selection, bubble, insertion, merge, quick = [], [], [], [], []

    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # Skip header
        for row in reader:
            sizes.append(int(row[0]))
            selection.append(int(row[1]))
            bubble.append(int(row[2]))
            insertion.append(int(row[3]))
            merge.append(int(row[4]))
            quick.append(int(row[5]))

    # Plot for each dataset
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, selection, marker='o', label='Selection Sort')
    plt.plot(sizes, bubble, marker='s', label='Bubble Sort')
    plt.plot(sizes, insertion, marker='^', label='Insertion Sort')
    plt.plot(sizes, merge, marker='*', label='Merge Sort')
    plt.plot(sizes, quick, marker='x', label='Quick Sort')

    plt.title(f'Input Size vs Steps Count ({title})')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Steps Count')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"graph_{title}.png", dpi=300, bbox_inches="tight")
    plt.show()
```
