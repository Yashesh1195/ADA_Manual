```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# File names and titles
data_files = {
    'random.txt': 'Random Data',
    'ascending.txt': 'Ascending Data',
    'descending.txt': 'Descending Data'
}

# Loop through each data file
for file_name, title in data_files.items():
    sizes = []
    selection, bubble, insertion, merge, quick = [], [], [], [], []

    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # Skip header
        for row in reader:
            sizes.append(int(row[0]))
            selection.append(int(row[1]))
            bubble.append(int(row[2]))
            insertion.append(int(row[3]))
            merge.append(int(row[4]))
            quick.append(int(row[5]))

    # Create a numpy array for the x-axis positions
    x = np.arange(len(sizes))
    bar_width = 0.15

    # Plot for each dataset
    plt.figure(figsize=(10, 6))
    plt.bar(x - 2*bar_width, selection, width=bar_width, label='Selection Sort')
    plt.bar(x - bar_width, bubble, width=bar_width, label='Bubble Sort')
    plt.bar(x, insertion, width=bar_width, label='Insertion Sort')
    plt.bar(x + bar_width, merge, width=bar_width, label='Merge Sort')
    plt.bar(x + 2*bar_width, quick, width=bar_width, label='Quick Sort')

    # # Add names at end of lines
    # plt.text(sizes[-1], selection[-1], 'Selection', fontsize=9, va='center', ha='left')
    # plt.text(sizes[-1], bubble[-1], 'Bubble', fontsize=9, va='center', ha='left')
    # plt.text(sizes[-1], insertion[-1], 'Insertion', fontsize=9, va='center', ha='left')
    # plt.text(sizes[-1], merge[-1], 'Merge', fontsize=9, va='center', ha='left')
    # plt.text(sizes[-1], quick[-1], 'Quick', fontsize=9, va='center', ha='left')

    plt.title(f'Input Size vs Steps Count ({title})')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Steps Count')
    # plt.xticks(x, sizes)  # Set x-axis ticks to input sizes
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"graph_{title}.png", dpi=300, bbox_inches="tight")
    plt.show()
```
