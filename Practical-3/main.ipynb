```python
import matplotlib.pyplot as plt

# Read data from file
inputs = []
best_case = []
avg_case = []
worst_case = []

with open("sequential_search_steps.txt", "r") as file:
    next(file)  # Skip header
    for line in file:
        parts = line.strip().split(",")
        if len(parts) >= 4:
            inputs.append(int(parts[0]))
            best_case.append(int(parts[1]))
            avg_case.append(int(parts[2]))
            worst_case.append(int(parts[3]))

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(inputs, best_case, 'go-', label="Best Case (O(1))")
plt.plot(inputs, avg_case, 'bo-', label="Average Case (O(n/2))")
plt.plot(inputs, worst_case, 'ro-', label="Worst Case (O(n))")

plt.title("Sequential Search Time Complexity Analysis")
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Steps")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('graph.png', dpi=300, bbox_inches="tight")
plt.show()
```
