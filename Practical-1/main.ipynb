```python
import matplotlib.pyplot as plt
import csv

# List to store data
inputs = []
loops_count = []
equation_count = []
recursion_count = []

# Reading from Java Output File
with open("steps_output.txt", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        inputs.append(int(row[0]))
        loops_count.append(int(row[1]))
        equation_count.append(int(row[2]))
        recursion_count.append(int(row[3]))

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(inputs, loops_count, marker='o', label = "Loop Method", color="blue")
plt.plot(inputs, equation_count, marker='s', label = "Equation Method", color="green")
plt.plot(inputs, recursion_count, marker='^', label = "Recursive Method", color="red")

plt.title("Inputs vs Steps for each Method")
plt.xlabel("Inputs")
plt.ylabel("Steps")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("graph.png", dpi=300, bbox_inches="tight")
plt.show()
```
